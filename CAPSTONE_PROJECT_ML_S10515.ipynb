```python
# PARSING OF THE REQUIRED COLUMNS


# Parsing of CPU column

import pandas as pd
import re

def simple_cpu_parse(cpu):
    if pd.isnull(cpu):
        return pd.Series([None, None, None])

    brand = cpu.split()[0]

    # Extract GHz speed
    speed_match = re.search(r"([\d.]+)\s*GHz", cpu)
    speed = float(speed_match.group(1)) if speed_match else None

    # Identify CPU series
    cpu_lower = cpu.lower()
    if "i3" in cpu_lower:
        series = "i3"
    elif "i5" in cpu_lower:
        series = "i5"
    elif "i7" in cpu_lower:
        series = "i7"
    elif "i9" in cpu_lower:
        series = "i9"
    elif "ryzen 3" in cpu_lower:
        series = "Ryzen 3"
    elif "ryzen 5" in cpu_lower:
        series = "Ryzen 5"
    elif "ryzen 7" in cpu_lower:
        series = "Ryzen 7"
    elif "ryzen 9" in cpu_lower:
        series = "Ryzen 9"
    elif "celeron" in cpu_lower:
        series = "Celeron"
    elif "pentium" in cpu_lower:
        series = "Pentium"
    elif "xeon" in cpu_lower:
        series = "Xeon"
    else:
        series = "Other"

    return pd.Series([brand, series, speed])

# Apply the parser to the DataFrame
df[["Cpu_Brand", "Cpu_Series", "Cpu_SpeedGHz"]] = df["Cpu"].apply(simple_cpu_parse)
print(df.columns)
```

```python
# Parsing of the GPU Column

def parse_gpu(gpu):
  if pd.isnull(gpu):
    return pd.Series([None,None])
  brand=gpu.split()[0]
  if "AMD" in gpu:
    series=re.search(r"AMD\s+(.+)",gpu)
  elif "Intel" in gpu:
    series=re.search(r"Intel\s+(.+)",gpu)
  elif "Nvidia" in gpu:
    series=re.search(r"NVIDIA\s+(.+)",gpu)
  else:
    series=None
  series_name = series.group(1) if series else "Other"
  return pd.Series([brand,series_name])
df[["Gpu_Brand","Gpu_Series"]]=df["Gpu"].apply(parse_gpu)
print(df.columns)
```

```python
import re

def parse_memory(mem):
    if pd.isnull(mem):
        return pd.Series([None, None, None, None])

    mem = mem.upper().replace("TB", "000").replace("GB", "")

    # Extract sizes
    sizes = re.findall(r"\d+", mem)
    sizes = [int(s) for s in sizes]

    # Extract types
    types = []
    if "SSD" in mem:
        types.append("SSD")
    if "HDD" in mem:
        types.append("HDD")
    if "EMMC" in mem:
        types.append("eMMC")

    # Pad with None if less than 2 values
    while len(sizes) < 2:
        sizes.append(None)
    while len(types) < 2:
        types.append(None)

    return pd.Series([sizes[0], types[0], sizes[1], types[1]])

# Apply to DataFrame
df[["Primary_Memory_Size", "Primary_Memory_Type", "Secondary_Memory_Size", "Secondary_Memory_Type"]] = df["Memory"].apply(parse_memory)
print(df.columns)
```

The model performs reasonably well overall, but it's likely less accurate for lesser-known brands because:

*  **Does the brand of the laptop significantly influence its price?**

Yes. Brand (Company) was among the selected features, and it's included in the model through one-hot encoding. This means the model has learned to associate different brands with different typical price ranges - confirming that brand has a notable effect on laptop pricing.



*  **How well does the model perform on laptops with high-end specifications compared to budget laptops?**

From the model’s predictions and error analysis:

The model performs better on budget and mid-range laptops.

On high-end laptops, the errors tend to be higher because there are fewer high-end examples in the data.

High-end pricing may involve brand premiums or design factors not included in my features.

To improve this, I could segment the evaluation (e.g., test on high-price vs low-price laptops separately).

Limited dataset variety.

Prices change over time, and the dataset may not reflect the latest trends.


Despite these, the model gives a useful estimate of price based on hardware.
